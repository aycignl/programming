```python
# plot popular mathematical functions 
import numpy as np
import matplotlib.pyplot as plt

# plot linear function
plt.plot([1, 2, 3, 4, 5], ':')
plt.xlabel('x-axes')
plt.ylabel('y-axes')
plt.title("Linear function")

data = np.linspace(0, 10, 100)
fig = plt.figure()
ax = plt.axes()

plt.plot(data, np.sin(data), '--', color = 'green', label = 'sin(data)')
plt.plot(data, np.cos(data), 'o', color = '#FFDD44', label = 'cos(data)')

ax.set(xlim = (0, 10), ylim = (-1, 1), 
      xlabel = 'data', ylabel = 'func(data)',
      title = "Trigonometric functions");

#plt.axis('equal')
plt.legend() # add label

fig.savefig('function.png')
```
